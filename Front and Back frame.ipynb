```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

data = pd.read_excel('/Users/nicol/Documents/python xls files/Sample Front and back frame.xlsx')

# links en rechts samen
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1,1,1, projection='3d')
for i in np.unique(data['color_l']):
    ax.plot(data['x1_l'][data['color_l'] == i], 
            data['y1_l'][data['color_l'] == i], 
            data['z1_l'][data['color_l'] == i], '-o')
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
```
