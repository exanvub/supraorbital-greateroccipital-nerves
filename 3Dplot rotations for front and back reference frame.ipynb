```python
import pandas as pd
import numpy as np
import os 
os.chdir('/Users/nicolas/Library/CloudStorage/OneDrive-VrijeUniversiteitBrussel/Onderzoek/n supraorbi occipitalis major/python xlsx files')


data_f  = pd.read_excel('197 supraorb python.xlsx')
data_b  = pd.read_excel('Mapping occip 197.xlsx')
data_fb = pd.read_excel('schedel 197.xlsx')
data_names = pd.read_excel('Sample Output names.xlsx')

# local reference frame for the back of the skull
PMR = np.array(data_b.iloc[0, np.array([2,3,4])]).reshape(1,3)
POE = np.array(data_b.iloc[1, np.array([2,3,4])]).reshape(1,3)
PML = np.array(data_b.iloc[2, np.array([2,3,4])]).reshape(1,3)
ob  = PMR
i   = (PML - PMR)/np.linalg.norm(PML - PMR)
h   = (POE - PMR)/np.linalg.norm(POE - PMR)
j   = np.cross(i, h)/np.linalg.norm(np.cross(i, h))
k   = np.cross(i, j)
rb  = np.array([i, j, k]).reshape(3,3)

# local reference frame for the front of the skull
HOL = np.array(data_f.iloc[0, np.array([2,3,4])]).reshape(1,3)
BON = np.array(data_f.iloc[1, np.array([2,3,4])]).reshape(1,3)
HOR = np.array(data_f.iloc[2, np.array([2,3,4])]).reshape(1,3)
of  = HOR
i   = (HOL - HOR)/np.linalg.norm(HOL - HOR)
h   = (BON - HOR)/np.linalg.norm(BON - HOR)
j   = np.cross(i, h)/np.linalg.norm(np.cross(i, h))
k   = np.cross(i, j)
rf  = np.array([i, j, k]).reshape(3,3)

# local reference frames for front and back of the skull measured together
HOL = np.array(data_fb.iloc[0, np.array([2,3,4])]).reshape(1,3)
BON = np.array(data_fb.iloc[1, np.array([2,3,4])]).reshape(1,3)
HOR = np.array(data_fb.iloc[2, np.array([2,3,4])]).reshape(1,3)
OF  = HOR
i   = (HOL - HOR)/np.linalg.norm(HOL - HOR)
h   = (BON - HOR)/np.linalg.norm(BON - HOR)
j   = np.cross(i, h)/np.linalg.norm(np.cross(i, h))
k   = np.cross(i, j)
RF  = np.array([i, j, k]).reshape(3,3)
PML = np.array(data_fb.iloc[3, np.array([2,3,4])]).reshape(1,3)
POE = np.array(data_fb.iloc[4, np.array([2,3,4])]).reshape(1,3)
PMR = np.array(data_fb.iloc[5, np.array([2,3,4])]).reshape(1,3)
OB  = PMR
i   = (PML - PMR)/np.linalg.norm(PML - PMR)
h   = (POE - PMR)/np.linalg.norm(POE - PMR)
j   = np.cross(i, h)/np.linalg.norm(np.cross(i, h))
k   = np.cross(i, j)
RB  = np.array([i, j, k]).reshape(3,3)

del i, j, k, h

def displace_x_to(x, r, R, o, O):
    
        # relative position to o (in global reference frame 1)
    LOC = x - o
    
    # coordiantes of LOC in local reference frame
    loc = np.dot(r, LOC.transpose()).transpose() 
    
    # rotate loc to global reference frame 2 and add O to it
    LOC2 = O + np.dot(R.transpose(), loc.transpose()).transpose()

    return LOC2

# nerves to rotate
# Back Left
Nocc_uit_semispin = np.array(data_b.iloc[3, np.array([2,3,4])]).reshape(1,3) 
Nocc_uit_aponeur  = np.array(data_b.iloc[4, np.array([2,3,4])]).reshape(1,3) 
split_med_lat_bun = np.array(data_b.iloc[5, np.array([2,3,4])]).reshape(1,3) 
med_tak_mid_prox  = np.array(data_b.iloc[7, np.array([2,3,4])]).reshape(1,3)
med_tak_mid       = np.array(data_b.iloc[8, np.array([2,3,4])]).reshape(1,3)
med_tak_mid_dist  = np.array(data_b.iloc[9, np.array([2,3,4])]).reshape(1,3)
med_tak_dist      = np.array(data_b.iloc[10, np.array([2,3,4])]).reshape(1,3)
lat_tak_split     = np.array(data_b.iloc[12, np.array([2,3,4])]).reshape(1,3)
lat_tak_med_mid   = np.array(data_b.iloc[14, np.array([2,3,4])]).reshape(1,3)
lat_tak_med_dist  = np.array(data_b.iloc[15, np.array([2,3,4])]).reshape(1,3)
lat_tak_lat_mid   = np.array(data_b.iloc[17, np.array([2,3,4])]).reshape(1,3)
lat_tak_lat_dist  = np.array(data_b.iloc[18, np.array([2,3,4])]).reshape(1,3)

# Back Right
Nocc_uit_semispin_R = np.array(data_b.iloc[3, np.array([7,8,9])]).reshape(1,3) 
Nocc_uit_aponeur_R  = np.array(data_b.iloc[4, np.array([7,8,9])]).reshape(1,3) 
split_med_lat_bun_R = np.array(data_b.iloc[5, np.array([7,8,9])]).reshape(1,3) 
med_tak_mid_prox_R  = np.array(data_b.iloc[7, np.array([7,8,9])]).reshape(1,3)
med_tak_mid_R       = np.array(data_b.iloc[8, np.array([7,8,9])]).reshape(1,3)
med_tak_mid_dist_R  = np.array(data_b.iloc[9, np.array([7,8,9])]).reshape(1,3)
med_tak_dist_R      = np.array(data_b.iloc[10, np.array([7,8,9])]).reshape(1,3)
lat_tak_split_R     = np.array(data_b.iloc[12, np.array([7,8,9])]).reshape(1,3)
lat_tak_med_mid_R   = np.array(data_b.iloc[14, np.array([7,8,9])]).reshape(1,3)
lat_tak_med_dist_R  = np.array(data_b.iloc[15, np.array([7,8,9])]).reshape(1,3)
lat_tak_lat_mid_R   = np.array(data_b.iloc[17, np.array([7,8,9])]).reshape(1,3)
lat_tak_lat_dist_R  = np.array(data_b.iloc[18, np.array([7,8,9])]).reshape(1,3)

# Front left, but accualy the right nerve
Sulc_supra_orb    = np.array(data_f.iloc[3, np.array([2,3,4])]).reshape(1,3) 
Nsupraorb_split   = np.array(data_f.iloc[4, np.array([2,3,4])]).reshape(1,3) 
MEDIAAL_mid_splitsing_mid = np.array(data_f.iloc[5, np.array([2,3,4])]).reshape(1,3) 
MEDIAAL_mid               = np.array(data_f.iloc[6, np.array([2,3,4])]).reshape(1,3) 
MEDIAAL_mid_dist_mid      = np.array(data_f.iloc[7, np.array([2,3,4])]).reshape(1,3) 
mediale_tak_dist  = np.array(data_f.iloc[8, np.array([2,3,4])]).reshape(1,3) 
INTER_mid_splitsing_mid = np.array(data_f.iloc[10, np.array([2,3,4])]).reshape(1,3) 
INTER_mid         = np.array(data_f.iloc[11, np.array([2,3,4])]).reshape(1,3) 
INTER_mid_dist_mid = np.array(data_f.iloc[12, np.array([2,3,4])]).reshape(1,3) 
inter_tak_dist    = np.array(data_f.iloc[13, np.array([2,3,4])]).reshape(1,3) 
LATERAAL_mid_splitsing_mid = np.array(data_f.iloc[15, np.array([2,3,4])]).reshape(1,3) 
LATERAAL_mid      = np.array(data_f.iloc[16, np.array([2,3,4])]).reshape(1,3) 
LATERAAL_mid_dist_mid = np.array(data_f.iloc[17, np.array([2,3,4])]).reshape(1,3) 
laterale_tak_dist = np.array(data_f.iloc[18, np.array([2,3,4])]).reshape(1,3) 

# Front Right, but accualy the left nerve
Sulc_supra_orb_R    = np.array(data_f.iloc[3, np.array([7,8,9])]).reshape(1,3) 
Nsupraorb_split_R   = np.array(data_f.iloc[4, np.array([7,8,9])]).reshape(1,3) 
MEDIAAL_mid_splitsing_mid_R = np.array(data_f.iloc[5, np.array([7,8,9])]).reshape(1,3) 
MEDIAAL_mid_R               = np.array(data_f.iloc[6, np.array([7,8,9])]).reshape(1,3) 
MEDIAAL_mid_dist_mid_R      = np.array(data_f.iloc[7, np.array([7,8,9])]).reshape(1,3) 
mediale_tak_dist_R  = np.array(data_f.iloc[8, np.array([7,8,9])]).reshape(1,3) 
INTER_mid_splitsing_mid_R = np.array(data_f.iloc[10, np.array([7,8,9])]).reshape(1,3) 
INTER_mid_R         = np.array(data_f.iloc[11, np.array([7,8,9])]).reshape(1,3) 
INTER_mid_dist_mid_R = np.array(data_f.iloc[12, np.array([7,8,9])]).reshape(1,3) 
inter_tak_dist_R    = np.array(data_f.iloc[13, np.array([7,8,9])]).reshape(1,3) 
LATERAAL_mid_splitsing_mid_R = np.array(data_f.iloc[15, np.array([7,8,9])]).reshape(1,3) 
LATERAAL_mid_R      = np.array(data_f.iloc[16, np.array([7,8,9])]).reshape(1,3) 
LATERAAL_mid_dist_mid_R = np.array(data_f.iloc[17, np.array([7,8,9])]).reshape(1,3) 
laterale_tak_dist_R = np.array(data_f.iloc[18, np.array([7,8,9])]).reshape(1,3) 

# Rotaties
# Back Left
Nocc_uit_semispin2 = displace_x_to(Nocc_uit_semispin, rb, RB, ob, OB)
Nocc_uit_aponeur2  = displace_x_to(Nocc_uit_aponeur, rb, RB, ob, OB) 
split_med_lat_bun2 = displace_x_to(split_med_lat_bun, rb, RB, ob, OB) 
med_tak_mid_prox2  = displace_x_to(med_tak_mid_prox, rb, RB, ob, OB)
med_tak_mid2       = displace_x_to(med_tak_mid, rb, RB, ob, OB)
med_tak_mid_dist2  = displace_x_to(med_tak_mid_dist, rb, RB, ob, OB)
med_tak_dist2      = displace_x_to(med_tak_dist, rb, RB, ob, OB)
lat_tak_split2     = displace_x_to(lat_tak_split, rb, RB, ob, OB)
lat_tak_med_mid2   = displace_x_to(lat_tak_med_mid, rb, RB, ob, OB)
lat_tak_med_dist2  = displace_x_to(lat_tak_med_dist, rb, RB, ob, OB)
lat_tak_lat_mid2   = displace_x_to(lat_tak_lat_mid, rb, RB, ob, OB)
lat_tak_lat_dist2  = displace_x_to(lat_tak_lat_dist, rb, RB, ob, OB)

# Back Right
Nocc_uit_semispin_R2 = displace_x_to(Nocc_uit_semispin_R, rb, RB, ob, OB)
Nocc_uit_aponeur_R2  = displace_x_to(Nocc_uit_aponeur_R, rb, RB, ob, OB) 
split_med_lat_bun_R2 = displace_x_to(split_med_lat_bun_R, rb, RB, ob, OB) 
med_tak_mid_prox_R2  = displace_x_to(med_tak_mid_prox_R, rb, RB, ob, OB)
med_tak_mid_R2       = displace_x_to(med_tak_mid_R, rb, RB, ob, OB)
med_tak_mid_dist_R2  = displace_x_to(med_tak_mid_dist_R, rb, RB, ob, OB)
med_tak_dist_R2      = displace_x_to(med_tak_dist_R, rb, RB, ob, OB)
lat_tak_split_R2     = displace_x_to(lat_tak_split_R, rb, RB, ob, OB)
lat_tak_med_mid_R2   = displace_x_to(lat_tak_med_mid_R, rb, RB, ob, OB)
lat_tak_med_dist_R2  = displace_x_to(lat_tak_med_dist_R, rb, RB, ob, OB)
lat_tak_lat_mid_R2   = displace_x_to(lat_tak_lat_mid_R, rb, RB, ob, OB)
lat_tak_lat_dist_R2  = displace_x_to(lat_tak_lat_dist_R, rb, RB, ob, OB)

# Front left, but accualy the right nerve
Sulc_supra_orb2    = displace_x_to(Sulc_supra_orb, rf, RF, of, OF)
Nsupraorb_split2   = displace_x_to(Nsupraorb_split, rf, RF, of, OF)
MEDIAAL_mid_splitsing_mid2 = displace_x_to(MEDIAAL_mid_splitsing_mid, rf, RF, of, OF)
MEDIAAL_mid2                = displace_x_to(MEDIAAL_mid, rf, RF, of, OF)
MEDIAAL_mid_dist_mid2       = displace_x_to(MEDIAAL_mid_dist_mid, rf, RF, of, OF)
mediale_tak_dist2           = displace_x_to(mediale_tak_dist, rf, RF, of, OF)
INTER_mid_splitsing_mid2    = displace_x_to(INTER_mid_splitsing_mid, rf, RF, of, OF)
INTER_mid2                  = displace_x_to(INTER_mid, rf, RF, of, OF)
INTER_mid_dist_mid2         = displace_x_to(INTER_mid_dist_mid, rf, RF, of, OF)
inter_tak_dist2             = displace_x_to(inter_tak_dist, rf, RF, of, OF)
LATERAAL_mid_splitsing_mid2 = displace_x_to(LATERAAL_mid_splitsing_mid, rf, RF, of, OF)
LATERAAL_mid2               = displace_x_to(LATERAAL_mid, rf, RF, of, OF)
LATERAAL_mid_dist_mid2      = displace_x_to(LATERAAL_mid_dist_mid, rf, RF, of, OF)
laterale_tak_dist2          = displace_x_to(laterale_tak_dist, rf, RF, of, OF)

#Front Right, but accualy the left nerve
Sulc_supra_orb_R2    = displace_x_to(Sulc_supra_orb_R, rf, RF, of, OF)
Nsupraorb_split_R2   = displace_x_to(Nsupraorb_split_R, rf, RF, of, OF)
MEDIAAL_mid_splitsing_mid_R2 = displace_x_to(MEDIAAL_mid_splitsing_mid_R, rf, RF, of, OF)
MEDIAAL_mid_R2                = displace_x_to(MEDIAAL_mid_R, rf, RF, of, OF)
MEDIAAL_mid_dist_mid_R2       = displace_x_to(MEDIAAL_mid_dist_mid_R, rf, RF, of, OF)
mediale_tak_dist_R2           = displace_x_to(mediale_tak_dist_R, rf, RF, of, OF)
INTER_mid_splitsing_mid_R2    = displace_x_to(INTER_mid_splitsing_mid_R, rf, RF, of, OF)
INTER_mid_R2                  = displace_x_to(INTER_mid_R, rf, RF, of, OF)
INTER_mid_dist_mid_R2         = displace_x_to(INTER_mid_dist_mid_R, rf, RF, of, OF)
inter_tak_dist_R2             = displace_x_to(inter_tak_dist_R, rf, RF, of, OF)
LATERAAL_mid_splitsing_mid_R2 = displace_x_to(LATERAAL_mid_splitsing_mid_R, rf, RF, of, OF)
LATERAAL_mid_R2               = displace_x_to(LATERAAL_mid_R, rf, RF, of, OF)
LATERAAL_mid_dist_mid_R2      = displace_x_to(LATERAAL_mid_dist_mid_R, rf, RF, of, OF)
laterale_tak_dist_R2          = displace_x_to(laterale_tak_dist_R, rf, RF, of, OF)


data = np.array([PML,
                      POE, 
                      PMR, 
                      HOL, 
                      BON,
                      HOR, 
                      Nocc_uit_semispin2, 
                      Nocc_uit_aponeur2, 
                      split_med_lat_bun2,
                      split_med_lat_bun2,
                      med_tak_mid_prox2,
                      med_tak_mid2,
                      med_tak_mid_dist2,
                      med_tak_dist2,
                      split_med_lat_bun2,
                      lat_tak_split2,
                      lat_tak_split2,
                      lat_tak_med_mid2,
                      lat_tak_med_dist2,
                      lat_tak_split2,
                      lat_tak_lat_mid2,
                      lat_tak_lat_dist2,
                      Sulc_supra_orb2,
                      Nsupraorb_split2,
                      MEDIAAL_mid_splitsing_mid2,
                      MEDIAAL_mid2,
                      MEDIAAL_mid_dist_mid2,
                      mediale_tak_dist2,
                      Nsupraorb_split2,
                      INTER_mid_splitsing_mid2,
                      INTER_mid2,
                      INTER_mid_dist_mid2,
                      inter_tak_dist2,
                      Nsupraorb_split2,
                      LATERAAL_mid_splitsing_mid2,
                      LATERAAL_mid2,
                      LATERAAL_mid_dist_mid2,
                      laterale_tak_dist2,
                    Nocc_uit_semispin_R2, 
                    Nocc_uit_aponeur_R2, 
                    split_med_lat_bun_R2,
                    split_med_lat_bun_R2,
                    med_tak_mid_prox_R2,
                    med_tak_mid_R2,
                    med_tak_mid_dist_R2,
                    med_tak_dist_R2,
                    split_med_lat_bun_R2,
                    lat_tak_split_R2,
                    lat_tak_split_R2,
                    lat_tak_med_mid_R2,
                    lat_tak_med_dist_R2,
                    lat_tak_split_R2,
                    lat_tak_lat_mid_R2,
                    lat_tak_lat_dist_R2,
                    Sulc_supra_orb_R2,
                    Nsupraorb_split_R2,
                    MEDIAAL_mid_splitsing_mid_R2,
                    MEDIAAL_mid_R2,
                    MEDIAAL_mid_dist_mid_R2,
                    mediale_tak_dist_R2,
                    Nsupraorb_split_R2,
                    INTER_mid_splitsing_mid_R2,
                    INTER_mid_R2,
                    INTER_mid_dist_mid_R2,
                    inter_tak_dist_R2,
                    Nsupraorb_split_R2,
                    LATERAAL_mid_splitsing_mid_R2,
                    LATERAAL_mid_R2,
                    LATERAAL_mid_dist_mid_R2,
                    laterale_tak_dist_R2

                                                 ]).reshape(70,3)
# Labels and colors:
label_l = []
label_l = list(data_names['label_l'])

color_l = []
color_l = list(data_names['color_l'])



# Make dataframe
df = pd.DataFrame(data)
df['color_l'] = color_l # adding a collum
df['label_l'] = label_l # adding a collum

# rename collums
df.columns =['x1_l', 'y1_l', 'z1_l', 'color_l', 'label_l']


# 3D plot

%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

data = df

fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1,1,1, projection='3d')
for i in np.unique(data['color_l']):
    ax.plot(data['x1_l'][data['color_l'] == i], 
            data['y1_l'][data['color_l'] == i], 
            data['z1_l'][data['color_l'] == i], '-o')

ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')

# export to excel
#df.to_excel('Sample_rotations_plot3.xlsx', index=False, header=True)
```
