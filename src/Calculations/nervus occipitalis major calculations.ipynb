```python
import numpy as np

def distance_point_to_line_segment(point, line_start, line_end):
    line_vector = line_end - line_start
    point_vector = point - line_start
    t = np.dot(point_vector, line_vector) / np.dot(line_vector, line_vector)
    t = np.clip(t, 0, 1)
    nearest_point = line_start + t * line_vector
    distance = np.linalg.norm(nearest_point - point)
    return distance

# Define points
L_pm = np.array(data_b.loc['processus mastoideus links', ['x1_l', 'y1_l','z1_l']])
R_pm = np.array(data_b.loc['processus mastoideus rechts', ['x1_l', 'y1_l','z1_l']])
PROTO = np.array(data_b.loc['protuberantia occiptalis externa', ['x1_l', 'y1_l','z1_l']])

L_apo       = np.array(data_b.loc['n occipitalis uittreden aponeurose', ['x1_l', 'y1_l','z1_l']])
R_apo       = np.array(data_b.loc['n occipitalis uittreden aponeurose', ['x1_r', 'y1_r','z1_r']])

L_semi     = np.array(data_b.loc['n occipitalis uittreden semispinalis', ['x1_l', 'y1_l','z1_l']])
R_semi     = np.array(data_b.loc['n occipitalis uittreden semispinalis', ['x1_r', 'y1_r','z1_r']])

L_ML_split = np.array(data_b.loc['splitsing mediale laterale bundel', ['x1_l', 'y1_l','z1_l']])
R_ML_split = np.array(data_b.loc['splitsing mediale laterale bundel', ['x1_r', 'y1_r','z1_r']])

L_med_dist = np.array(data_b.loc['med tak dist', ['x1_l', 'y1_l','z1_l']])
R_med_dist = np.array(data_b.loc['med tak dist', ['x1_r', 'y1_r','z1_r']])

L_med_mid  = np.array(data_b.loc['med tak mid', ['x1_l', 'y1_l','z1_l']])
R_med_mid  = np.array(data_b.loc['med tak mid', ['x1_r', 'y1_r','z1_r']])

L_med_mid_dist = np.array(data_b.loc['med tak mid dist', ['x1_l', 'y1_l','z1_l']])
R_med_mid_dist = np.array(data_b.loc['med tak mid dist', ['x1_r', 'y1_r','z1_r']])

L_med_mid_prox = np.array(data_b.loc['med tak mid prox', ['x1_l', 'y1_l','z1_l']])
R_med_mid_prox = np.array(data_b.loc['med tak mid prox', ['x1_r', 'y1_r','z1_r']])

L_lat_split = np.array(data_b.loc['lat tak splitsing ', ['x1_l', 'y1_l','z1_l']])
R_lat_split = np.array(data_b.loc['lat tak splitsing ', ['x1_r', 'y1_r','z1_r']])

L_lat_med_dist = np.array(data_b.loc['lat tak med dist', ['x1_l', 'y1_l','z1_l']])
R_lat_med_dist = np.array(data_b.loc['lat tak med dist', ['x1_r', 'y1_r','z1_r']])

L_lat_med_mid = np.array(data_b.loc['lat tak med mid', ['x1_l', 'y1_l','z1_l']])
R_lat_med_mid = np.array(data_b.loc['lat tak med mid', ['x1_r', 'y1_r','z1_r']])

L_lat_lat_dist = np.array(data_b.loc['lat tak lat dist', ['x1_l', 'y1_l','z1_l']])
R_lat_lat_dist = np.array(data_b.loc['lat tak lat dist', ['x1_r', 'y1_r','z1_r']])

L_lat_lat_mid = np.array(data_b.loc['lat tak lat mid', ['x1_l', 'y1_l','z1_l']])
R_lat_lat_mid = np.array(data_b.loc['lat tak lat mid', ['x1_r', 'y1_r','z1_r']])

# Calculate middle point between L_pm and R_pm
mid_point = (L_pm + R_pm) / 2

# Calculate distance
L_sagi_apo = distance_point_to_line_segment(L_apo, mid_point, PROTO)
L_sagi_semi = distance_point_to_line_segment(L_semi, mid_point, PROTO)
L_sagi_ML_split = distance_point_to_line_segment(L_ML_split, mid_point, PROTO)
L_sagi_lat_split = distance_point_to_line_segment(L_lat_split, mid_point, PROTO)
L_sagi_med_dist = distance_point_to_line_segment(L_med_dist, mid_point, PROTO)
L_sagi_lat_med_dist = distance_point_to_line_segment(L_lat_med_dist, mid_point, PROTO)
L_sagi_lat_lat_dist = distance_point_to_line_segment(L_lat_lat_dist, mid_point, PROTO)

R_sagi_apo = distance_point_to_line_segment(R_apo, mid_point, PROTO)
R_sagi_semi = distance_point_to_line_segment(R_semi, mid_point, PROTO)
R_sagi_ML_split = distance_point_to_line_segment(R_ML_split, mid_point, PROTO)
R_sagi_lat_split = distance_point_to_line_segment(R_lat_split, mid_point, PROTO)
R_sagi_med_dist = distance_point_to_line_segment(R_med_dist, mid_point, PROTO)
R_sagi_lat_med_dist = distance_point_to_line_segment(R_lat_med_dist, mid_point, PROTO)
R_sagi_lat_lat_dist = distance_point_to_line_segment(R_lat_lat_dist, mid_point, PROTO)

```
